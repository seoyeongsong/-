```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
date_data = pd.date_range(start="20110603", end="20190809")
date_data = date_data.strftime("%Y%m%d")
date_data = date_data.to_list()
```

## 날짜별 뉴스 main 페이지에서 특정 키워드 검색

```python
start_url = "https://media.daum.net/ranking/bestreply?regDate="
```

```python
keyword = "무임"
keyword2 = "노인"
```

```python
for date in date_data:
    main_url = start_url + date
    
    # url 접속
    req = requests.get(main_url)
    soup = BeautifulSoup(req.text, "lxml")
    
    # 기사 제목과 내용이 있는 태그 추출
    data = soup.find_all("a", {"class" : "link_txt"})
    
    # 키워드 검색
    i = 0
    for line in data[:50]:
        if not keyword in line.text:
            i +=1
            pass
        else:
            if keyword2 not in line.text:
                i +=1
                pass
            else:
                
                print(date)
                print(line.text)
                print(data[i]['href'])
                i += 1
```

## 키워드를 포함하는 기사의 제목과 링크 csv파일로 저장하기

```python
date_of_article = []
title = []
link = []

for date in date_data:
    main_url = start_url + date
    
    # url 접속
    req = requests.get(main_url)
    soup = BeautifulSoup(req.text, "lxml")
    
    # 기사 제목과 내용이 있는 태그 추출
    data = soup.find_all("a", {"class" : "link_txt"})
    
    # 키워드 검색
    i = 0
    for line in data[:50]:
        if not keyword in line.text:
            pass
        else:
            if keyword2 not in line.text:
                pass
            else:
                
#                 print(date)
#                 print(line.text)
#                 print(data[i]['href'])
                
                t = line.text
                l = data[i]["href"]
                date_of_article.append(date)
                title.append(t)
                link.append(l)
                i += 1
print(len(title))
print(len(link))
```

```python
df = pd.DataFrame()
df['날짜'] = date_of_article
df['제목'] = title
df['링크'] = link
df.to_csv(keyword+"&"+keyword2+".csv", index=False, encoding = "euc-kr")
```
